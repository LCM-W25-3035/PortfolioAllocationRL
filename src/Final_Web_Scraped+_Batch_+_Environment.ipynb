```python
import yfinance as yf
import pandas as pd
import time

# Function to fetch historical data for a list of tickers
def fetch_all_historical_data(ticker_list, start_date, end_date, requests_per_minute=100):
    data = []
    delay = 60 / requests_per_minute  # Calculate delay to respect rate limits

    for i, ticker in enumerate(ticker_list):
        try:
            print(f"Fetching data for: {ticker}")
            # Download historical data for the ticker
            ticker_data = yf.download(ticker, start=start_date, end=end_date, progress=False)

            # Check if data is empty
            if ticker_data.empty:
                print(f"Warning: No data found for {ticker}")
                continue

            # Add a 'tic' column with the ticker symbol
            ticker_data['tic'] = ticker
            ticker_data.reset_index(inplace=True)  # Reset index to bring 'Date' as a column

            # Select only the required columns
            ticker_data = ticker_data[['Date', 'Close', 'High', 'Low', 'Open', 'Volume', 'tic']]

            # Add day of the week
            ticker_data['day'] = pd.to_datetime(ticker_data['Date']).dt.dayofweek

            # Rename columns to lowercase
            ticker_data.columns = ['date', 'close', 'high', 'low', 'open', 'volume', 'tic', 'day']

            # Append to the data list
            data.append(ticker_data)

        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")

        # Delay to respect the request limit
        if i < len(ticker_list) - 1:
            time.sleep(delay)

    # Combine all data into a single DataFrame
    if not data:
        print("Error: No data found for any of the specified tickers.")
        return pd.DataFrame()
    else:
        combined_data = pd.concat(data, ignore_index=True)

        # Sort by 'date' and 'tic'
        sorted_data = combined_data.sort_values(by=['date', 'tic'], ascending=[True, True])
        return sorted_data

# Example usage
ticker_list = FIRST_SET_TICKERS
start_date = "2004-01-01"
end_date = "2024-12-31"

# Fetch historical data
historical_data = fetch_all_historical_data(
    ticker_list=ticker_list,
    start_date=start_date,
    end_date=end_date,
    requests_per_minute=100
)

# Verify the shape and columns
print("DataFrame columns:", historical_data.columns.tolist())
print("DataFrame shape:", historical_data.shape)
print("\nFirst few rows:")
print(historical_data.head())
```

```python
import pandas as pd
import numpy as np

class StockDataIterator:
    def __init__(self, file_path, lookback_window=10, chunk_size=100000):
        self.file_path = file_path
        self.lookback_window = lookback_window
        self.chunk_size = chunk_size

        # Read the first row to get column names dynamically
        with open(self.file_path) as f:
            first_line = f.readline().strip().split(',')

        # Ensure expected columns exist (without 'volume')
        expected_columns = {'date', 'close', 'high', 'low', 'tic'}
        missing_columns = expected_columns - set(first_line)
        if missing_columns:
            raise ValueError(f"Missing columns in CSV: {missing_columns}")

        self.columns = first_line  # Dynamically assign columns
        print("Initializing iterator...")
        self._initialize_metadata()

        self.current_chunk = 0
        self.data_buffer = pd.DataFrame()

    def _initialize_metadata(self):
        """Fast row count without loading data"""
        with open(self.file_path) as f:
            self.total_rows = sum(1 for _ in f) - 1  # Exclude header

        self.num_chunks = (self.total_rows + self.chunk_size - 1) // self.chunk_size
        print(f"Total rows: {self.total_rows:,}, Chunks: {self.num_chunks}")

    def _load_next_chunk(self):
        """Optimized chunk loading"""
        if self.current_chunk >= self.num_chunks:
            return None

        chunk = pd.read_csv(
            self.file_path,
            skiprows=1 + self.current_chunk * self.chunk_size,
            nrows=self.chunk_size,
            names=self.columns,
            usecols=['date', 'tic', 'close', 'high', 'low'],  # Omit 'volume'
            dtype={
                'close': 'float32',
                'high': 'float32',
                'low': 'float32',
                'tic': 'category'
            },
            parse_dates=['date'],
            engine='c'
        )

        if chunk.empty:
            return None

        chunk = chunk.sort_values(['date', 'tic'])
        self.current_chunk += 1
        return chunk

    def __iter__(self):
        self.current_chunk = 0
        self.data_buffer = pd.DataFrame(columns=['date', 'tic', 'close', 'high', 'low'])
        return self

    def __next__(self):
        """Main processing loop with validation"""
        while True:
            # Ensure buffer has enough dates
            while self.data_buffer.empty or self.data_buffer['date'].nunique() < self.lookback_window + 2:
                chunk = self._load_next_chunk()
                if chunk is None:
                    break
                self.data_buffer = pd.concat([self.data_buffer, chunk], ignore_index=True, sort=False)

            if self.data_buffer.empty or self.data_buffer['date'].nunique() < self.lookback_window:
                raise StopIteration

            # Get sorted unique dates
            dates = np.sort(self.data_buffer['date'].unique())
            window_dates = dates[:self.lookback_window]

            # Filter valid stocks
            window_mask = self.data_buffer['date'].isin(window_dates)
            tic_counts = self.data_buffer[window_mask].groupby('tic').size()
            valid_tics = tic_counts[tic_counts == self.lookback_window].index

            if not valid_tics.empty:
                valid_mask = window_mask & self.data_buffer['tic'].isin(valid_tics)
                window_data = self.data_buffer[valid_mask]
                prepared = self._prepare_features(window_data)

                if prepared:
                    # Drop the oldest date to move forward
                    self.data_buffer = self.data_buffer[~self.data_buffer['date'].isin([window_dates[0]])]
                    return prepared

            # Drop the oldest date and continue
            self.data_buffer = self.data_buffer[~self.data_buffer['date'].isin([window_dates[0]])]

    def _prepare_features(self, window_data):
        """Feature calculation"""
        if window_data.empty:
            return []

        features = []
        grouped = window_data.groupby('tic', observed=False)  # Fixed observed=False

        for tic, group in grouped:
            if len(group) != self.lookback_window:
                continue

            close = group['close'].values.astype('float32')
            high = group['high'].values.astype('float32')
            low = group['low'].values.astype('float32')

            features.append({
                'tic': tic,
                'close_prices': close,
                'returns': np.diff(close, prepend=close[0]) / close[0],
                'log_returns': np.log(close / np.roll(close, 1)),
                'price_range': (high - low) / close  # Keep price range feature
            })

        return features

def demonstrate_iterator():
    print("Creating optimized iterator...")
    iterator = StockDataIterator(
        '/content/historical_data.csv',
        lookback_window=10,
        chunk_size=10000
    )

    print("\nProcessing first 3 valid batches:")
    count = 0
    for batch in iterator:
        if count >= 3:
            break
        if batch:
            print(f"\nBatch {count+1}:")
            print(f"Stocks: {len(batch)}")
            print(f"Features for {batch[0]['tic']}:")
            for key, val in batch[0].items():
                print(f" - {key}: {val[:3]}...")
            count += 1

if __name__ == "__main__":
    demonstrate_iterator()
```

```python
!pip install gym
!pip install finrl
!pip install stable-baselines3
!pip install pandas

import gym
from gym import spaces

class StockTradingEnv(gym.Env):
    def __init__(self, data_iterator):
        super(StockTradingEnv, self).__init__()

        self.data_iterator = data_iterator
        self.initial_cash = 10000.0

        # Define observation space for flattened features
        num_features = 4  # close_prices, returns, log_returns, price_range
        self.lookback_window = data_iterator.lookback_window

        # Flatten the observation space
        self.observation_space = spaces.Box(
            low=-np.inf,
            high=np.inf,
            shape=(self.lookback_window * num_features,),
            dtype=np.float32
        )

        self.action_space = spaces.Discrete(3)  # Buy (0), Hold (1), Sell (2)
        self.iterator = iter(self.data_iterator)

    def _get_observation(self):
        if not self.current_features:
            return np.zeros(self.observation_space.shape)

        # Flatten and concatenate all features
        features = self.current_features[0]  # Take first stock's features
        obs = np.concatenate([
            features['close_prices'],
            features['returns'],
            features['log_returns'],
            features['price_range']
        ])
        return obs.astype(np.float32)

    def reset(self):
        self.cash = self.initial_cash
        self.stock_held = 0
        self.iterator = iter(self.data_iterator)  # Reset iterator

        try:
            self.current_features = next(self.iterator)
        except StopIteration:
            self.current_features = None

        return self._get_observation()

    def step(self, action):
        if self.current_features is None:
            return self._get_observation(), 0, True, {}

        current_price = self.current_features[0]['close_prices'][-1]
        prev_portfolio_value = self.cash + self.stock_held * current_price

        # Execute action
        if action == 0 and self.cash >= current_price:  # Buy
            self.stock_held += 1
            self.cash -= current_price
        elif action == 2 and self.stock_held > 0:  # Sell
            self.stock_held -= 1
            self.cash += current_price

        try:
            self.current_features = next(self.iterator)
            new_price = self.current_features[0]['close_prices'][-1]
        except StopIteration:
            return self._get_observation(), 0, True, {}

        current_portfolio_value = self.cash + self.stock_held * new_price
        reward = (current_portfolio_value - prev_portfolio_value) / prev_portfolio_value  # Percentage return

        return self._get_observation(), reward, False, {}

    def render(self):
        if self.current_features:
            current_price = self.current_features[0]['close_prices'][-1]
            portfolio_value = self.cash + self.stock_held * current_price
            print(f"Cash: {self.cash:.2f}, Stocks: {self.stock_held}, "
                  f"Price: {current_price:.2f}, Portfolio: {portfolio_value:.2f}")
```
