```python
import yfinance as yf
import pandas as pd
import time
```

```python
def fetch_tickers_info(symbols):
    # List to store ticker info dictionaries
    data = []

    # Iterate over the symbols
    for symbol in symbols:
        try:
            # Fetch the ticker info
            ticker = yf.Ticker(symbol)
            info = ticker.info

            # Append the info dictionary to the list
            data.append(info)
            print(f"Fetched info for {symbol}")

            # Delay to avoid rate limiting
            time.sleep(1)
        except Exception as e:
            print(f"Error fetching data for {symbol}: {e}")

    # Convert the list of dictionaries into a DataFrame
    df = pd.DataFrame(data)
    return df
```

```python
# Load the Excel file
df = pd.read_csv('TSX stocks.csv')

# Convert the 'Symbol' column to an array (list)
symbols_array = df['Symbol'].tolist()

# Display the array
print(symbols_array)
```
